```python
import random
import numpy as np
import pandas as pd
import trading_env
import tushare as ts
import datetime
```

```python
df1 = ts.get_tick_data('600000',date='2017-06-06')
df2 = ts.get_tick_data('600000',date='2017-06-07')
df3 = ts.get_tick_data('600000',date='2017-06-08')
# df4 = ts.get_tick_data('600000',date='2017-06-09')
# df5 = ts.get_tick_data('600000',date='2017-06-10')

df=pd.concat([df1,df2])
df=pd.concat([df,df3])
# df=pd.concat([df,df4])
# df=pd.concat([df,df5])
```

```python
df.change=df.change.apply(lambda x :0 if x=='--' else x )
df.index=list(range(len(df)))
```

```python
df['serial_number']=df.index
```

```python
df = df.rename(columns={'time': 'datetime'})
```

```python
df.datetime=pd.to_datetime(df.datetime)
```

```python
# import random
# import numpy as np
# import pandas as pd
# import trading_env

# df = pd.read_hdf('dataset/SGXTW.h5', 'STW')

env = trading_env.make(obs_data_len=256, step_len=128,
                      df=df, fee=0.01, max_position=5, deal_col_name='price', 
                      feature_names=['price', 'change', 
                                     'volume','amount', ], 
                      fluc_div=100.0)

env.reset()
env.render()

state, reward, done, info = env.step(random.randrange(3))

### randow choice action and show the transaction detail
for i in range(500):
    print(i)
    state, reward, done, info = env.step(random.randrange(3))
    print(state, reward)
    env.render()
    if done:
        break
env.transaction_details
```
